Experment E1_1 (simple split) <br>
Hyperparameter choice for Experiment 1: 25 epoches

```python
e1_f3_r1 = EvaluationResultContainer.from_json("saved_data/eval_results/k_fold/E1_1_3fold_rep_0.json")
e1_f3_r2 = EvaluationResultContainer.from_json("saved_data/eval_results/k_fold/E1_1_3fold_rep_1.json")
e1_f3_r3 = EvaluationResultContainer.from_json("saved_data/eval_results/k_fold/E1_1_3fold_rep_2.json")


E1_fig,axs = mk_3fold_3reps_fig(e1_f3_r1*100, e1_f3_r2*100, e1_f3_r3*100,epoch_range=(5,105,5))
# zoom in on the interesting parts
tl_axs = axs[0:2]
sq_axs = axs[2:4]
[ax.set_ylim([0,1]) for ax in tl_axs]
[ax.set_ylim([0,2]) for ax in sq_axs]
axs[-1].set_ylim([0,2])
```
